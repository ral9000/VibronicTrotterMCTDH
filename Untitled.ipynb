```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Read the pickle file
with open('data/dabna_6s_5m_t500_dt=0.01.pkl', 'rb') as f:
    data = pickle.load(f)
    
# Display the type and structure of the data
print(f"Data type: {type(data)}")

# If it's a dictionary, show the keys
if isinstance(data, dict):
    print("\nKeys in the data:")
    for key in data.keys():
        print(f"- {key}: {type(data[key])}")
        
        # For numpy arrays or lists, show shape/length and a sample
        if isinstance(data[key], np.ndarray):
            print(f"  Shape: {data[key].shape}")
            print(f"  Sample: {data[key].flatten()[:5]}...")
        elif isinstance(data[key], list):
            print(f"  Length: {len(data[key])}")
            print(f"  Sample: {data[key][:5]}...")
            
# Basic visualization if the data contains arrays
def try_plot_data(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if isinstance(value, np.ndarray) and len(value.shape) <= 2:
                plt.figure(figsize=(10, 6))
                if len(value.shape) == 1 or value.shape[1] == 1:
                    plt.plot(value.flatten())
                    plt.title(f"Plot of {key}")
                elif value.shape[1] <= 10:  # Only plot if not too many columns
                    for i in range(value.shape[1]):
                        plt.plot(value[:, i], label=f"Column {i}")
                    plt.legend()
                    plt.title(f"Plot of {key}")
                plt.savefig(f"{key}_plot.png")
                print(f"Created plot for {key}")

try:
    try_plot_data(data)
    print("\nCheck for any generated plot images in your current directory.")
except Exception as e:
    print(f"Couldn't create plots: {e}")
```
